```python
train_listimages = []
train_listlabels = []
```

```python
test_listimages = []
#test_listlabels = []
```

```python
import cv2
```

```python
for i,imagepath in enumerate(train_listpaths):
  image = cv2.imread(imagepath)
  print(image.shape)
  image = cv2.resize(image,(32,32))
  image = image.flatten()
  label = imagepath.split("/")[-2]
  train_listlabels.append(label)
  train_listimages.append(image)
```

```python
for i,imagepath in enumerate(test_listpaths):
  image = cv2.imread(imagepath)
  print(image.shape)
  image = cv2.resize(image,(32,32))
  image = image.flatten()
  #label = imagepath.split("/")[-2]
  #train_listlabels.append(label)
  test_listimages.append(image)
```

```python
import numpy as np
```

```python
listlabels = np.array(train_listlabels)
listimages = np.array(train_listimages)
```

```python
test_listimages = np.array(test_listimages)
```

```python
listimages.shape
```

```python
test_listimages.shape
```

```python
listlabels
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
int_labels = lb.fit_transform(listlabels)
```

```python
int_labels
```

```python
lb.classes_
```

```python
from tensorflow.keras.utils import to_categorical
number_of_classes = 5

y = to_categorical(int_labels, number_of_classes)
```

```python
y[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(listimages, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(1024,input_dim=3072,activation='relu'))
model.add(Dense(512,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(5,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,epochs=100,batch_size=46,validation_data=(x_test,y_test))
```

```python
test_predict = model.predict(test_listimages)
```

```python
test_predict
```

```python
np.argmax(test_predict,axis=1)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
model.save("animals.h5")
```
